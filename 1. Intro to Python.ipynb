### Control the precision of printed numbers

Notice that the print statement above prints a seemingly arbitrary number of values after the decimal.

Yes, there is a default setting for this that you can change, but there is a better way.

You can use print() in combination with an f-string, which I demonstrate below.  

You can learn more about f-strings here:
https://www.geeksforgeeks.org/formatted-string-literals-f-strings-python/



```python
blargh = 1.12345

print(f'Blargh is: {blargh}')
print(f'Blargh is: {blargh:.1f}') # One value of precision
print(f'Blargh is: {blargh:.2f}') # Two values of precision
print(f'Blargh is: {blargh:.3f}') # Three values of precision
```

# Iteration through list-like objects using a for loop (iterables)

To iterate means to move through a list-like variable element by element.  Things that can be iterated are referred to as "iterable." Here are a few methods for interation using the *for* loop.

Note that in Python, the contents of a for loop require indentation of *one tab*.  **The indentation is primordial and at the core of Python.**  

```python
list_of_strings = ['alpha', 'beta', 'gamma', 'delta']

for i in list_of_strings:
  print(i)
```

You can use the range() function to index into specific values.  Below, we demonstrate how the variable x is controlled using for() and range().

Note that the start value is included in the range, but the stop value is excluded.

```python
print("Range (from 0) to 4 (excluded) ")
for x in range(4):
    print(x)

print("Range from 2 (included) to 6 (excluded) ")
for x in range(2,6):
    print(x)

print("Range from 1 (included) to 12 (excluded) by steps of 3 ")
for x in range(1,12,3):
    print(x)
```

Now, we will use for() and range() to index into a list.

```python
a_list = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

print('Even numbers:')
for i in range(0,10,2): # from 0 to 10 in steps of 2
  print( a_list[i])

print('\nEvery third number:') # \n is a linebreak!
for i in range(0,10,3):
  print( a_list[i])
```

### Need to keep track of where you are when iterating through a list?  There are at least two ways.

1) Assign a count variable that is incremented within the for loop.

2) You can use enumerate(). At each iteration/step through the list, enumerate() assigns both its index location in the list and the value at location.

```python
a_list = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']


print("Using a variable counter:")
count = 0;
for value in a_list: # note that the variable name "value" could be any other name.
  print(str(count) + ": " +value )
  count = count +1

# Again, "index" and "value" are arbitrary variable names, but the index will always be the first and the value the second
print("\nUsing enumerate():")
for index, value in enumerate(a_list):
  print(str(index) + ": " +value )

# Let me demonstrate the arbitray nature of those variable names.
# All that matters is that the index is assigned to the first variable, and the value to the second.
print("\nUsing enumerate() and different variable names:")
for chicken, cow in enumerate(a_list):
  print(str(chicken) + ": " +cow )
```

```python
def my_fun():
    print("You call this fun!??!")

my_fun()
```

This function returns the string.

```python
def square(x):
    return(x * x)

print( square(12) )
print( square(13) )
print( square(14) )
```

Funtions can be very helpful when automating tasks.

```python
def in_bedify(string_in):

    if string_in[-1] == '.': # if there the sentence ends in a "."
      string_in = string_in[:-1] # ...get everything but the last char

    string_out = string_in + ' in bed!'
    return string_out

# from http://www.fortunecookiemessage.com/
list_of_fortunes = ['Go confidently in the direction of your dreams.',
                    'Good things come to those who wait',
                    'A wise man once said, everything in its own time and place']

for i in list_of_fortunes:
  print( in_bedify(i) )
```

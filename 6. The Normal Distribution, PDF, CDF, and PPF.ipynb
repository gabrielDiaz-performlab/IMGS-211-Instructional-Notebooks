Let's plot the cumulative distribution function.

```python
N = 1001 # back to the full 1001
xVals = np.linspace(-5, 5, N)
normPDF = normStandardDist.pdf(xVals)

# Find the CDF for each value along the X axis
normCDF = normStandardDist.cdf(xVals)
```

```python

plt.plot(xVals, normPDF, label = 'PDF')
plt.plot(xVals, normCDF, label = 'CDF')

plt.title(f'Standard normal distribution. $\mu={mu}. \sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.axhline(color='black', linewidth=0.25)
plt.axvline(color='k', lw=0.25)
plt.show()
```

Let's find the area under the curve at a particular value of X:

```python
x = 1
CDF_at_X = normStandardDist.cdf(x)

print(f'p(x<{x}) = cdf({x}) = {normStandardDist.cdf(x):0.4f}')
print(f'p(x>{x}) = 1-cdf({x}) = {1-normStandardDist.cdf(x):0.4f}')
```

# What is the area within +/- one standard deviation from the mean?


```python

x1 = -1
x2 = 1

area_between = normStandardDist.cdf(x2) - normStandardDist.cdf(x1)

plt.plot(xVals, normPDF, label = 'PDF')
fill_range = np.linspace(-1, 1, 100)
plt.fill_between(fill_range, normStandardDist.pdf(fill_range),color='g',alpha=0.5)
plt.text(-5.2,0.38,'$\int_{-1}^{1}PDF(x) dx=$' + f'{area_between:.4f}',fontsize=18,ha='left')

plt.title(f'Standard normal distribution. $\mu={mu}. \sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.axhline(color='black', linewidth=0.25)
plt.axvline(color='k', lw=0.25)
plt.show()
```

# The percentile point function

```python
mu = 0
sigma = 1
normStandardDist = stats.norm(mu,sigma)
print(f'NormStandardDist = {normStandardDist}')

fraction_between = 0.95
zLow = (1-fraction_between) / 2
zHigh = fraction_between + zLow

lowerCl, upperCl = normStandardDist.ppf((zLow, zHigh))

print(f'{fraction_between}% of the data lies between {lowerCl:.3f} and {upperCl:.3f} zScores/StDev of the mean.')
```

What portion of the distribution containts 95% of the data?

```python
N = 1001 # back to the full 1001
xVals = np.linspace(-5, 5, N)

pdf_data = normStandardDist.pdf(xVals)
cdf_data = normStandardDist.cdf(xVals)

pct_data = np.linspace(0, 1, N)
ppf_data = normStandardDist.ppf(pct_data)

fig, axs = plt.subplots(1, 3, figsize=(20, 7))

plt.subplot(131)
plt.plot(xVals, pdf_data)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.xlim((-3,3))

plt.subplot(132)
plt.plot(xVals, cdf_data)
plt.title(f'Standard normal distribution. $\mu={mu}. \sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('cumulative probability of x')
plt.xlim((-3,3))
plt.ylim((0,1))

plt.subplot(133)
plt.plot(pct_data, ppf_data)
plt.xlabel('percent of data')
plt.ylabel('$\sigma$')
plt.axhline(color='black', linewidth=0.25)
plt.ylim((-3,3))
plt.xlim((0,1))


# plt.tight_layout(1.02)

plt.show()
```

Note that PPF does not return x values, but the location along if in units of standard deviation (or, zscores!). TO demonstrate this, let me plot the same distributions for a generalized normal distribution with a mu and std not equal to 1 and 0

```python
mu = 20
sigma = 3
genNormalDist = stats.norm(mu,sigma)
print(f'genNormalDist = {genNormalDist}')

N = 1001 # back to the full 1001
xVals = np.linspace(10, 30, N)

pdf_data = genNormalDist.pdf(xVals)
cdf_data = genNormalDist.cdf(xVals)

pct_data = np.linspace(0, 1, N)
ppf_data = genNormalDist.ppf(pct_data)

fig, axs = plt.subplots(1, 3, figsize=(20, 7))

plt.subplot(131)
plt.plot(xVals, pdf_data)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.xlim((10,30))

plt.subplot(132)
plt.plot(xVals, cdf_data)
plt.title(f'Standard normal distribution. $\mu={mu}. \sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('cumulative probability of x')
plt.xlim((10,30))
plt.ylim((0,1))

plt.subplot(133)
plt.plot(pct_data, ppf_data)
plt.xlabel('percent of data')
plt.ylabel('$\sigma$')
plt.axhline(color='black', linewidth=0.25)
plt.ylim((10,30))
plt.xlim((0,1))

plt.show()
```
